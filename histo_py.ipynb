```python
import pandas as pd
import numpy as np

# Generate a large dataset with 1,000,000 rows and 3 columns
np.random.seed(0)
data = np.random.randn(1000000, 3)

# Create a DataFrame
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

# Display the first few rows
df.head()
```

```python
# Calculate summary statistics
summary_stats = df.describe()

# Display summary statistics
print(summary_stats)
```

```python
import matplotlib.pyplot as plt

# Plot histograms of the columns
df.hist(bins=50, figsize=(10, 8))
plt.suptitle('Histograms of A, B, and C')
plt.show()
```
